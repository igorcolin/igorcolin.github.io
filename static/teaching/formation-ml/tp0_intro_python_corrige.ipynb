## Python en bref

Les briques de base : variables, listes, dictionnaires, boucles et fonctions.

```python
# une liste et un dictionnaire
notes = [12, 15, 9, 18, 7]
etudiant = {"nom": "Alice", "moyenne": 14.2}

print("nombre de notes :", len(notes))
print("max :", max(notes), "| min :", min(notes))
print(etudiant["nom"], "->", etudiant["moyenne"])

# compréhension de liste : les notes >= 10
recues = [n for n in notes if n >= 10]
print("notes >= 10 :", recues)
```

### Exercice 1
Complétez la fonction `moyenne` qui renvoie la moyenne d'une liste de nombres (sans utiliser `numpy`).

```python
def moyenne(valeurs):
    return sum(valeurs) / len(valeurs)

def moyenne_lente(valeurs):
    res = 0
    for val in valeurs:
        res = res + val
    return res / len(valeurs)

print("moyenne des notes        :", moyenne(notes))
print("moyenne des notes (lente):", moyenne_lente(notes))
```

## NumPy : calcul sur des tableaux

NumPy remplace les boucles par des opérations **vectorisées**, bien plus rapides et lisibles. Un tableau (`array`) se comporte comme une matrice.

```python
import numpy as np

a = np.array([1.0, 2.0, 3.0, 4.0])
M = np.arange(12).reshape(3, 4)   # matrice 3x4

print("a * 2     =", a * 2)            # opération élément par élément
print("a.mean()  =", a.mean())
print("M shape   :", M.shape)
print("colonne 0 :", M[:, 0])          # indexation : toutes les lignes, colonne 0
print("moyenne par colonne :", M.mean(axis=0))
```

### Exercice 2
**Standardisez** le tableau `x` : retranchez la moyenne et divisez par l'écart-type, en une seule expression vectorisée. Le résultat doit avoir une moyenne ~0 et un écart-type ~1.

```python
x = np.array([10.0, 12.0, 14.0, 16.0, 18.0])
x_std = (x - x.mean()) / x.std()
print("standardisé :", np.round(x_std, 2))
print("moyenne =", round(x_std.mean(), 3), "| écart-type =", round(x_std.std(), 3))
```

La vectorisation est aussi une affaire de vitesse. On compare une somme de carrés en boucle Python contre NumPy.

```python
import time

grand = np.random.rand(1_000_000)

t0 = time.time(); s1 = sum(v * v for v in grand); t1 = time.time()
s2 = np.sum(grand ** 2); t2 = time.time()

print(f"boucle Python : {t1 - t0:.3f} s")
print(f"NumPy         : {t2 - t1:.4f} s")
print("même résultat :", np.isclose(s1, s2))
```

## Matplotlib : visualiser

`matplotlib.pyplot` permet de tracer courbes et nuages de points.

```python
import matplotlib.pyplot as plt

t = np.linspace(0, 2 * np.pi, 200)
plt.plot(t, np.sin(t), label="sin")
plt.plot(t, np.cos(t), label="cos")
plt.legend(); plt.title("Deux courbes"); plt.xlabel("t"); plt.show()
```

## pandas : explorer un tableau de données

Un `DataFrame` est un tableau étiqueté (colonnes nommées), comme une feuille de calcul. On construit celui des iris (un jeu classique : 150 fleurs, 4 mesures, 3 espèces).

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df = iris.frame.drop(columns="target")   # les 4 mesures
df["espece"] = iris.target_names[iris.target]
print(df.shape)
df.head()
```

### Exercice 4
Calculez la **moyenne de chaque mesure, par espèce** (indice : `groupby("espece").mean(...)`). On ne garde que les colonnes numériques avec `numeric_only=True`.

```python
resume = df.groupby("espece").mean(numeric_only=True)
print(resume)
```

```python
# un aperçu visuel : longueur vs largeur des pétales, par espèce
for nom, sous in df.groupby("espece"):
    plt.scatter(sous["petal length (cm)"], sous["petal width (cm)"], s=15, label=nom)
plt.xlabel("longueur pétale"); plt.ylabel("largeur pétale")
plt.legend(); plt.title("Iris"); plt.show()
```

## Un premier modèle de bout en bout

On enchaîne maintenant la chaîne complète du ML supervisé : séparer train/test, entraîner, prédire, évaluer. C'est exactement le schéma repris dans les TP1 et TP2.

```python
from sklearn.model_selection import train_test_split

X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)
print("train :", X_train.shape[0], "| test :", X_test.shape[0])
```

### Exercice 5
Entraînez une `LogisticRegression` (mettez `max_iter=200`) sur le jeu d'entraînement, puis affichez son accuracy sur le test avec `.score(...)`.

```python
from sklearn.linear_model import LogisticRegression

modele = LogisticRegression(max_iter=200)
modele.fit(X_train, y_train)
acc = modele.score(X_test, y_test)
print(f"accuracy test : {acc:.3f}")
```

### Exercice 6
Affichez la prédiction du modèle pour les **trois premières fleurs du jeu de test**, à côté de leur vraie espèce.

```python
pred = modele.predict(X_test)
for i in range(3):
    print(f"prédit : {iris.target_names[pred[i]]:>12} | vrai : {iris.target_names[y_test.iloc[i]]}")
```
